```python
repeat_customer = df["card"].value_counts().head(5)
print(repeat_customer)
```

```python
plt.figure(figsize = (6,6))
ax = sns.barplot(x=repeat_customer.index, y=repeat_customer.values)
for i, value in enumerate(repeat_customer.values):
    ax.text(i, value + 0.5, str(value), ha='center', fontsize=12)
    plt.title("Top 5 Repeat Customer")
plt.xlabel("Customer Card")
plt.xticks(rotation=45)
plt.ylabel("Number of Transaction")
plt.show()
```

```python
top_5_customers = df["card"].value_counts().head(5).index
filtered_df = df[df["card"].isin(top_5_customers)]
customer_coffee_counts = filtered_df.groupby(["card", "coffee_name"]).size().reset_index(name="count")
print(customer_coffee_counts)
```

```python
fig, axes = plt.subplots(2, 3, figsize=(18, 12))
for i, customer in enumerate(top_5_customers):
    row, col = divmod(i, 3)  # Menentukan posisi subplot (baris, kolom)
    customer_data = customer_coffee_counts[customer_coffee_counts["card"] == customer]
    
    axes[row, col].pie(
        customer_data["count"],
        labels=customer_data["coffee_name"],
        autopct='%1.1f%%',
        startangle=140,
        colors=plt.cm.Paired.colors
    )
    
    axes[row, col].set_title(f"Customer: {customer}")
axes[1, 2].axis("off")

plt.tight_layout()
plt.show()
```
